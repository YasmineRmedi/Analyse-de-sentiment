```python
import pandas as pd
import re

import html
import unidecode

# Clean the text
def clean_text(text):
    if isinstance(text, str):  # V√©rifie si text est une cha√Æne de caract√®res
        text = re.sub(r'http\S+', '', str(text))  # Supprime les URLs
        text = re.sub(r'@\w+', '', str(text))      # Supprime les mentions
        text = re.sub(r'#\w+', '', str(text))      # Supprime les hashtags
    # Convertir le texte en minuscules
        text = text.lower()

    # Supprimer les balises HTML
        text = html.unescape(re.sub(r'<.*?>', '', text))

    # Supprimer les caract√®res non imprimables (non-ASCII)
        text = ''.join(char for char in text if char.isprintable())

    # Normaliser les caract√®res sp√©ciaux et accents
        text = unidecode.unidecode(text)
   
    # Supprime la ponctuation
        text = re.sub(r'[.,!:?"‚Äú‚Äù]', '', str(text))
    # Supprime les guillemets doubles vides
        text = text.replace('""', '')
    # Supprimer les nombres
        text = re.sub(r'\d+', '', text)
    # Supprimer les caract√®res sp√©ciaux
        text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Supprimer les espaces suppl√©mentaires
        text = re.sub(r'\s+', ' ', text)

    

    return text


# Apply the clean_text function to the 'Commentaire' column
commentaires['Commentaire'] = commentaires['Commentaire'].apply(clean_text)

# Remove duplicate comments
commentaires = commentaires.drop_duplicates(subset='Commentaire')
```

```python
# Supprimer les lignes vides du DataFrame
commentaires.dropna(inplace=True)
```
